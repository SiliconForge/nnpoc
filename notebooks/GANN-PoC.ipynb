```python
class individual:
    def __init__(self,neuronInterval=(2,10),inputs=1,outputs=1, debugFlag=False):
        self.debugFlag = debugFlag
        self.colorNodes = {0:"black",1:"red",2:"blue",3:"cyan",4:"yellow",5:"green",6:"pink"}
        self.activations = {0:lambda z: self.sigmoid(z),
                            1:lambda z: self.tanh(z),
                            2:lambda z: self.relu(z),
                            3:lambda z: self.elu(z),
                            4:lambda z: self.softplus(z),
                            5:lambda z: self.linear(z),
                            6:lambda z: self.inverter(z)}
        self.numInputs = inputs
        self.numOutputs = outputs
        self.numHidden = np.random.randint(*neuronInterval)
        self.totalNeurons = self.numHidden + self.numInputs + self.numOutputs
        self.adjecencyMatrix = self.generateAdjecencyMatrix()
        self.biasVector = self.initializeBias(self.adjecencyMatrix.shape[0])
        self.activationVector = self.initializeActivation(self.adjecencyMatrix.shape[0])
        self.weightMatrix = self.initializeWeights(np.random.random()*5)
        self.gene = self.initializeGene()
        if self.debugFlag: print("Individual successfully generated")
            
    def initializeBias(self,numBiases):
        biasVectorTmp = []
        for i in range(numBiases):
            biasVectorTmp.append(np.random.random()*10)
        return np.array(biasVectorTmp)
    
    #NZT-48
    def growNeuron(self):
        self.adjecencyMatrix=(np.insert(self.adjecencyMatrix,-self.numOutputs,np.zeros(self.adjecencyMatrix.shape[1]), axis=0))
        self.adjecencyMatrix=(np.insert(self.adjecencyMatrix,-self.numOutputs,np.zeros(self.adjecencyMatrix.shape[0]), axis=1))
        meanWeigths = self.weightMatrix.mean()
        self.weightMatrix=(np.insert(self.weightMatrix,-self.numOutputs,np.full(self.weightMatrix.shape[1],meanWeigths), axis=0))
        self.weightMatrix=(np.insert(self.weightMatrix,-self.numOutputs,np.full(self.weightMatrix.shape[0],meanWeigths), axis=1))
        self.biasVector = np.insert(self.biasVector,-self.numOutputs,self.biasVector.mean())
        self.activationVector = np.insert(self.activationVector,-self.numOutputs,np.random.randint(0,7))
    #Weed
    #def loseNeuron(self):
    
    def initializeActivation(self,numActivations):
        activationVectorTmp = []
        for i in range(numActivations):
            activationVectorTmp.append(np.random.choice(range(0,7),p=[1/7,1/7,1/7,1/7,1/7,1/7,1/7]))
        return activationVectorTmp
    
    def initializeGene(self):
        geneArray = []
        for i in range(self.adjecencyMatrix.shape[0]):
            connectionList = []
            weightList = []
            for j in range(self.adjecencyMatrix.shape[1]):
                if self.adjecencyMatrix[i,j] == 1:
                    connectionList.append(j)
                    weightList.append(self.weightMatrix[i,j])
            biasValue = self.biasVector[i]
            activationType = self.activationVector[i]
            geneArray.append([connectionList,weightList,activationType,biasValue,None])
        geneArray = np.array(geneArray)
        return geneArray
            
    def calculateOutputs(self,inputVector):
        for i in range(len(self.gene)):
            self.gene[i,4] = None
        for i in range(len(inputVector)):
            self.gene[i,4] = inputVector[i]
        for i in range(len(self.gene)):
            self.gene[i,4] = self.response(self.gene[i,0],i)
        return self.gene[len(self.gene)-self.numOutputs:,4]
        
    def softplus(self,x):
        return np.log(1+np.exp(x))
    
    def linear(self,x):
        return x
    
    def inverter(self,x):
        return -x

    def sigmoid(self,x):
        return 1/(1+np.exp(-x))
    
    def elu(self,x):
        if x <= 0:
            return (np.exp(x)-1)
        else:
            return x
    
    def tanh(self,x):
        return np.tanh(x)
    
    def relu(self,x):
        if x <= 0:
            return 0
        else:
            return x
    
    def response(self,indices,number):
        if self.gene[number,4] == None:
            answer = 0
            j = 0
            for i in indices:
                answer = answer + self.response(self.gene[i,0],i)*self.gene[number,1][j]
                j += 1
            return self.activations[self.gene[number,2]](answer+self.gene[number,3])
        else:    
            return self.gene[number,4]
        
    def generateAdjecencyMatrix(self,
                                hidden_connectedness = 0.5,
                                input_connectedness = 0.5,
                                output_connectedness = 0.5):
        retry = True
        counter = 0
        while retry:
            counter += 1
            retry = False
            adjecencyMatrixTmp = np.random.choice([0, 1],(self.totalNeurons,self.totalNeurons),p=[1-hidden_connectedness,hidden_connectedness])
            adjecencyMatrixTmp[self.totalNeurons-self.numOutputs:,:] = np.random.choice([0,1],(self.numOutputs,self.totalNeurons),p=[1-output_connectedness,output_connectedness])
            adjecencyMatrixTmp[:,:self.numInputs] = np.random.choice([0,1],(self.totalNeurons,self.numInputs),p=[1-input_connectedness,input_connectedness])
            adjecencyMatrixTmp[:self.numInputs,:] = np.zeros(adjecencyMatrixTmp[:self.numInputs,:].shape)
            adjecencyMatrixTmp[:,self.totalNeurons-self.numOutputs:] = np.zeros(adjecencyMatrixTmp[:,self.totalNeurons-self.numOutputs:].shape)
            for i in range(self.totalNeurons):
                for j in range(self.totalNeurons):
                    if i <= j:
                        adjecencyMatrixTmp[i,j] = 0
            retry = self.checkConsistencyAdjecency(adjecencyMatrixTmp)
        if self.debugFlag: 
            if counter % 100 == 1 : print("Iteration :" + str(counter))
        if self.debugFlag: print("Valid Topology found in "+str(counter)+" tries!")
        return adjecencyMatrixTmp
    
    def checkConsistencyAdjecency(self,testMatrix):
        retry = False
        for i in range(self.totalNeurons-self.numInputs):
            if (testMatrix[i+self.numInputs,:] == np.zeros(testMatrix[i+self.numInputs,:].shape)).all():
                retry = True
        for i in range(self.totalNeurons-self.numOutputs-self.numInputs):
            if (testMatrix[:,i+self.numInputs] == np.zeros(testMatrix[:,i+self.numInputs].shape)).all():
                retry = True
        return retry
    
    def mutateWeights(self):
        for i in range(self.weightMatrix.shape[0]):
            for j in range(self.weightMatrix.shape[1]):
                if np.random.random() <= 0.1:
                    self.weightMatrix[i,j] += (np.random.random()-0.5)*10
                
    def mutateAdjecency(self):
        retry = True
        while retry:
            adjecencyMatrixTmp = self.adjecencyMatrix.copy()
            for i in range(self.adjecencyMatrix.shape[0]):
                for j in range(self.adjecencyMatrix.shape[1]):
                    if np.random.random() <= 0.1 and i>j:
                        adjecencyMatrixTmp[i,j] = 2-(2**adjecencyMatrixTmp[i,j]) #(flips between 0 and N for any N with expression (N+1)-(N+1)**x where x is the flipping number(flipee?!))
            retry = self.checkConsistencyAdjecency(adjecencyMatrixTmp)
        self.adjecencyMatrix = adjecencyMatrixTmp.copy()
    
    def mutateBias(self):
        for i in range(len(self.biasVector)):
            if np.random.random() <= 0.1:
                self.biasVector[i] += (np.random.random()-0.5)*10
        
    def mutateActivation(self):
        for i in range(len(self.activationVector)):
            if np.random.random() <= 0.1:
                self.activationVector[i] = np.random.choice(range(0,7),p=[1/7,1/7,1/7,1/7,1/7,1/7,1/7])
                    
    def mutateIndividual(self):
        if np.random.random() < 0.1:
            self.growNeuron()
        self.mutateWeights()
        self.mutateBias()
        self.mutateAdjecency()
        self.mutateActivation()
        self.gene = self.initializeGene()
    
    def initializeWeights(self,weightRange):
        weightMatrixTmp = np.zeros(self.adjecencyMatrix.shape)
        for i in range(self.adjecencyMatrix.shape[0]):
            for j in range(self.adjecencyMatrix.shape[1]):
                weightMatrixTmp[i,j] = np.random.normal(0,weightRange)
        return weightMatrixTmp
    
    def visualizeTopology(self):
        graphvizString = "digraph{rankdir = LR;graph [ dpi = 50 ];"
        for i in range(len(self.adjecencyMatrix)):
            for j in range(len(self.adjecencyMatrix)):
                if self.adjecencyMatrix[i,j] == 1:
                    graphvizString += str(j) +\
                                      "->" + \
                                      str(i) +\
                                      "[label=" +\
                                      str(f"{self.weightMatrix[i,j]:0.2f}") +\
                                      " color=" +\
                                      self.colorNodes[self.gene[i,2]] +\
                                      "];" 
        graphvizString += "}"
        return graphvizString
```

```python
arrayIndividuals = []
originalMan = individual((5,6),1,1)
for i in range(10):
    tmpIndividual = copy.deepcopy(originalMan)
    tmpIndividual.mutateIndividual()
    arrayIndividuals.append(tmpIndividual)
plt.figure(figsize=(12,12))
x = np.linspace(-20,20,1000)
y = np.zeros(x.shape)
for j in range(len(arrayIndividuals)):
    for i in range(len(x)):
        y[i] = arrayIndividuals[j].calculateOutputs([x[i]])
    display(graphviz.Source(arrayIndividuals[j].visualizeTopology()))
    plt.plot(x,y)
    plt.ylim([-10,10])
```

```python
def mutateInduvidual(original,weightRate=0.1,biasRate=0.1):
    clone = copy.deepcopy(original)
    for i in range(len(original.gene)):
        if np.random.random() < biasRate:
            clone.gene[i,3] += ((np.random.random()-0.5)*10)
        for j in range(len(original.gene[i,1])):
            if np.random.random() < weightRate:
                clone.gene[i,1][j] += ((np.random.random()-0.5)*10)  #mutate weights  
    return clone
```
